#TESTING PYKALMAN (SAMPLING, EM, ETC)

```python
from pykalman import KalmanFilter
import numpy.ma as ma
import numpy as np
matplotlib.rcParams['figure.figsize'] = (15,10)
from scipy import linalg
from numpy import dot, linalg
from numpy.linalg import inv 
import pylab as pl

import copy
```

#SAMPLING PYKALMAN

```python
dim = 2
n_samples = 500

t_cov = [[1.2, 1.3], [.1, 1.4]]
t_obs = [[2.2, .3], [1.1, 2.4]]
# 0.2*eye(dim)

kf_sample = KalmanFilter(transition_covariance = t_cov , observation_covariance = t_obs)
samples = kf_sample.sample(n_samples)

hidden = samples[0]
obs = samples[1]

```

```python
pylab.figure()
pylab.plot(obs ,color = 'y')
pylab.plot(hidden ,color = 'r')
pylab.show()
```

#EM PYKALMAN

```python
kf_em = KalmanFilter(em_vars = ['transition_covariance', 'observation_covariance'], n_dim_obs=2, n_dim_state=2)
kf_em.em(obs, em_vars=['transition_covariance', 'observation_covariance'])

print(kf_em.transition_covariance)
print(kf_em.observation_covariance)


# Filtering
filtered = kf_em.filter(obs)
hidden_filtered = filtered[0]
```

```python
#Plotting
pylab.figure()
pylab.plot(hidden_filtered ,color = 'r',label='filtered hidden states')
pylab.plot(hidden ,color = 'b',label='real hidden')


pylab.plot(obs ,color = 'y',label='obs')
pylab.show()

err_filtered = hidden_filtered - hidden
print('Standard deviation of error for filtered series: ', err_filtered.std())

pylab.plot(err_filtered ,color = 'g',label='error filtered')
pylab.show()



```

#FILLING GAPS BY RUNNING EM AND SAMPLING (NOT SO PROMISING)

```python
mask = []
for n in arange(n_samples):
    if (n > 150 and n < 200):
        mask.append([1,0])
    else:
        if (n > 300 and n < 350):
            mask.append([0,1])
        else:
            mask.append([0,0])

masked_obs = ma.masked_array(obs, mask=mask)
```

```python
pylab.figure()
pylab.plot(masked_obs ,color = 'y')
pylab.show()
```

```python
kf = KalmanFilter(transition_covariance = 0.2*eye(dim), observation_covariance = 1.5*eye(dim))

samps_1 = kf.sample(49, initial_state=masked_obs.data[151], random_state=23)[1]
samps_2 = kf.sample(49, initial_state=masked_obs.data[301], random_state=0)[1]
```

```python
for n in arange(49):
    if (masked_obs.mask[151+n][0] == True):
        masked_obs.mask[151+n][0] = False
        masked_obs.data[151+n][0] = samps_1[n][0]
        

for n in arange(49):
    if (masked_obs.mask[301+n][1] == True):
        masked_obs.mask[301+n][1] = False
        masked_obs.data[301+n][1] = samps_2[n][1]
```

```python
pylab.figure()
pylab.plot(masked_obs ,color = 'y')
pylab.show()
```

```python
filtered = kf_sample.filter(masked_obs)
hidden_filtered = filtered[0]
```

```python
pylab.figure()
pylab.plot(hidden_filtered ,color = 'g',label='filtered hidden states')
pylab.plot(masked_obs ,color = 'y',label='real hidden states')
pylab.legend()
pylab.show()
```
